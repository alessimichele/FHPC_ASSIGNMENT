```python
def txt_to_pgm(input_txt, output_pgm):
    # Read the content of the input text file
    with open(input_txt, 'r') as file:
        lines = file.readlines()

    # Convert '1' to '255' in the content
    modified_lines = [line.replace('1', '255') for line in lines]

    # Write the modified content to the output PGM file
    with open(output_pgm, 'w') as file:
        # Write PGM header
        file.write("P5\n")
        file.write("# generated by\n")
        file.write("# Alessi Michele and Carollo Marco\n")
        file.write("{} {}\n".format(len(modified_lines[0].split()), len(modified_lines)))
        file.write("255\n")

        # Write modified content
        file.writelines(modified_lines)

# Example usage:
txt_input_file = 'bichitos.txt'
pgm_output_file = 'init_00500.pgm'
txt_to_pgm(txt_input_file, pgm_output_file)
```
